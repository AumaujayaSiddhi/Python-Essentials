```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# line plot
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y, color='r', marker='*',markeredgecolor='b', markersize=12)
plt.show()
```

```python
#histogram plot
x = [1,1,2,3,4,4,4,5,5,6,7,7,7,7,7,10]

plt.hist(x)
plt.show()
```

Like above you can find different kinds of plot like boxplot, violin plot, are plot, scatter plots, pie charts etc...
